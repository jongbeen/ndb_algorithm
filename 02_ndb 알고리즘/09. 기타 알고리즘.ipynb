### I. 소수판별

#### 1-1. Brute-Force

```python
def is_prime_number(x):
    for i in range(2,x):
        if x%i==0:
            return False
    return True

print(is_prime_number(5))
print(is_prime_number(4))
```

##### >>2부터 X-1까지 모든 연산을 수행하므로 시간복잡도는 O(X)

#### 1-2. 약수(제곱근) 을 활용한 성능향상


```python
import math

def is_prime_number(x):
    for i in range(2,int(math.sqrt(x))+1):
        if x%i==0:
            return False
    return True

print(is_prime_number(4))
print(is_prime_number(7))
```


##### >> 시간 복잡도를 절반가량 줄임

#### 1-3. 다수의 소수판별 (에라토스테네스의 체 알고리즘)



##### >> 5까지만 제거해도 소수를 제외한 거의 모든수가 제외됨

```python
import math

n = 1000 # 2부터 1000까지 모든 수에 대해 소수 판별
array = [True for i in range(n+1)]

for i in range(2,int(math.sqrt(n))+1):
    if array[i] == True:
        j = 2
        while i*j <= n:
            array[i*j] = False
            j += 1

for i in range(2,n+1):
    if array[i]:
        print(i, end=' ')
```


##### >>다만 메모리가 비효율적이다.

### II. 투포인터



##### >> 선형적으로 동작 [O(N)], 완전탐색 X [O(N^2)]


##### >> 이렇게 s,e를 증가시키며 N까지 count

```python
n = 5
m = 5
data = [1,2,3,2,5]

count = 0
interval_sum = 0
end = 0

for start in range(n):
    while interval_sum <m and end <n:
        interval_sum += data[end]
        end+=1
    if interval_sum == m:
        count +=1
    interval_sum -= data[start]
    
print(count)
```

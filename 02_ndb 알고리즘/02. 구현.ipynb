#### 문제1. 상하좌우


- 1<=N<=100

##### 내 답

```python
N = int(input())
plan = list(input().split())

cur_pos = [0,0]
for dir in plan:
    if dir == 'L':
        if cur_pos[1] == 0:
            continue
        else :
            cur_pos[1] -= 1 
    elif dir == 'R':
        if cur_pos[1] == N-1:
            continue
        else:
            cur_pos[1] += 1
    elif dir == 'U':
        if cur_pos[0] == 0:
            continue
        else:
            cur_pos[0] -=1
    elif dir == 'D':
        if cur_pos[0] == N-1:
            continue
        else:
            cur_pos[0] +=1
print(cur_pos[0]+1, cur_pos[1]+1)
```

##### 답안

```python
n = int(input())
x,y = 1,1
plans = input().split()

dx = [0,0,-1,1]
dy = [-1,1,0,0]
move_types = ['L','R','U','D']

for plan in plans:
    for i in range(len(move_types)):
        if plan == move_types[i]:
            nx = x + dx[i]
            ny = y + dy[i]
            
    if nx<1 or ny<1 or nx > n or nx > n:
        continue
    x,y = nx, ny
print(x,y)
```

#### 문제 2. 시각


##### 내 답

```python
def nH3M():
    return 60
    
def nHnM():
    cnt = 0
    for i in range(0,60):
        sec = str(i)
        if '3' in sec:
            cnt += 1
    return cnt

N = int(input())
count = 0
for hours in range(0,N+1):
    hour = str(hours)
    if '3' not in hour:
        for minutes in range(0, 60):
            minute = str(minutes)
            if '3' not in minute:
                count += nHnM()
            else:
                count += nH3M()
    else:
        count+=3600
print(count)
```

##### 답안

```python
h = int(input())
count = 0
for i in range(h+1):
    for j in range(60):
        for k in range(60):
            if '3' in str(i) + str(j) + str(k):
                count+=1
print(count)
```

#### 문제 3. 왕실의 나이트


##### 내답

```python
pos = input()
x,y = ord(pos[0]) - ord('a'), int(pos[1])
N = 8

dx = [-2,-2,-1,-1,1,1,2,2]
dy = [-1,1,-2,2,-2,2,-1,1]
count = 0

for nx,ny in zip(dx,dy):
    nx,ny = x+nx, y+ny
    
    if 1<=nx<=N and 1<=ny<=N:
        count+=1
print(count)
```

#### 문제 4. 문자열 재정렬


```python
string = list(input())
subA, subN = [],[]
for i in string:
    if 'A'<=i<='Z':
        subA.append(i)
    else:
        i = int(i)
        subN.append(i)
subA.sort()
subN.sort()
ans = ''.join(subA)
for i in subN:
    ans+=str(i)
print(ans)
```

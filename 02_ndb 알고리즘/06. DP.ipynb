##### Top-down  DP 

```python
d = [0]*100
def fibo(x):
    if x==1 or x==2:
        return 1
    if d[x] != 0:
        return d[x]
    d[x] = fibo(x-1) + fibo(x-2)
    return d[x]
print(fibo(99))
```

##### Bottom-up  DP 

```python
d = [0]*100
d[1] = 1
d[2] = 1
n = 99

for i in range(3,n+1):
    d[i] = d[i-1] + d[i-2]
print(d[n])
```

```python
N = int(input())
table = list(map(int,input().split()))
table.insert(0,0)
dp = [0] * (N+1)
dp[1] = table[1]
for i in range(2,N+1):
    dp[i] = max(table[i]+table[i-2],dp[i-1])
print(dp[N])
```

```python
from collections import deque

div = [5,3,2]

def bfs(x):
    dp,cnt = [False] * 30001,0
    dp[x] = True
    queue = deque([(x,cnt)])
    while queue:
        x,cnt = queue.popleft()
        if x == 1:
            return cnt
        for i in div:
            if x % i == 0:
                mod = x//i
                if not dp[mod]:
                    queue.append((mod,cnt+1))
        if x-1 >= 1:
            queue.append((x-1,cnt+1))
                

N = int(input())
print(bfs(N))
```

```python
MAX = 1e10

N = int(input())
dp = [0] * 30001
dp[1] = 0
dp[2] = 1
dp[3] = 1
dp[4] = 2
dp[5] = 1

if N<=5:
    print(dp[N])
else:
    for i in range(6, N+1):
        first,second,third,fourth = MAX,MAX,MAX,MAX
        if i%5 == 0:
            mod = i//5
            first = dp[mod]+1
        if i%3 == 0:
            mod = i//3
            second = dp[mod]+1
        if i%2 == 0:
            mod = i//2
            third = dp[mod]+1
        fourth = dp[i-1]+1
        dp[i] = min(first,second,third,fourth)
    print(dp[N])
```
